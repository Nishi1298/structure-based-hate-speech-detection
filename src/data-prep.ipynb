# Data Preparation

```python
import csv
import re
import numpy as np
import pandas as pd
from tqdm import tqdm

DATA_PATH = "../data/"
```

```python
DATA = DATA_PATH + "labeled_data.csv"
```

```python
def readData(path):
    data = []
    with open(path,'r') as file:
        data = [x for x in csv.reader(file, delimiter=',')]
    return data
```

```python
raw = readData(DATA)
raw = raw[1:]
```

```python
dd = []
with open(DATA, 'r') as file:
    for line in file:
       dd.append(line.strip())
```
